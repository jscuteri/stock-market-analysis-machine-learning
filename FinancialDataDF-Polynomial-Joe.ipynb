```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
import scipy.stats as st
from scipy.stats import linregress
import datetime
from datetime import datetime, timedelta
from datetime import timezeone
```

```python
### Import Sector Data

SP500_Material= pd.read_csv("Resources/Sectors/XLB.csv")
SP500_Energy= pd.read_csv("Resources/Sectors/XLE.csv")
SP500_Finance= pd.read_csv("Resources/Sectors/XLF.csv")
SP500_Industrial= pd.read_csv("Resources/Sectors/XLI.csv")
SP500_Tech= pd.read_csv("Resources/Sectors/XLK.csv")
SP500_Staples= pd.read_csv("Resources/Sectors/XLP.csv")
SP500_Utilities= pd.read_csv("Resources/Sectors/XLU.csv")
SP500_Health= pd.read_csv("Resources/Sectors/XLV.csv")
SP500_Discretionary= pd.read_csv("Resources/Sectors/XLY.csv")
```

```python
### Sector Data
SP500_Material= SP500_Material.drop(columns=['High','Low', 'Adj Close'])
SP500_Material= SP500_Material.rename(columns={'Open': 'SPMAT_Open',
                        'Close':'SPMAT_Close',
                        'Volume':'SPMAT_Volume'})

##Energy
SP500_Energy= SP500_Energy.drop(columns=['High','Low', 'Adj Close'])
SP500_Energy= SP500_Energy.rename(columns={'Open': 'SPENERGY_Open',
                        'Close':'SPENERGY_Close',
                        'Volume':'SPENERGY_Volume'})
##Finance
SP500_Finance= SP500_Finance.drop(columns=['High','Low', 'Adj Close'])
SP500_Finance= SP500_Finance.rename(columns={'Open': 'SPFINANCE_Open',
                        'Close':'SPFINANCE_Close',
                        'Volume':'SPFINANCE_Volume'})

## Industrial

SP500_Industrial=SP500_Industrial.drop(columns=['High','Low', 'Adj Close'])
SP500_Industrial= SP500_Industrial.rename(columns={'Open': 'SPINDUS_Open',
                        'Close':'SPINDUS_Close',
                        'Volume':'SPINDUS_Volume'})
##Tech
SP500_Tech=SP500_Tech.drop(columns=['High','Low', 'Adj Close'])
SP500_Tech= SP500_Tech.rename(columns={'Open': 'SPTech_Open',
                        'Close':'SPTech_Close',
                        'Volume':'SPTech_Volume'})

##Staples
SP500_Staples=SP500_Staples.drop(columns=['High','Low', 'Adj Close'])
SP500_Staples= SP500_Staples.rename(columns={'Open': 'SPStaples_Open',
                        'Close':'SPStaples_Close',
                        'Volume':'SPStaples_Volume'})

#Utilities
SP500_Utilities=SP500_Utilities.drop(columns=['High','Low', 'Adj Close'])
SP500_Utilities= SP500_Utilities.rename(columns={'Open': 'SPUTIL_Open',
                        'Close':'SPUTIL_Close',
                        'Volume':'SPUTIL_Volume'})

##Health
SP500_Health=SP500_Health.drop(columns=['High','Low', 'Adj Close'])
SP500_Health= SP500_Health.rename(columns={'Open': 'SPHealth_Open',
                        'Close':'SPHealth_Close',
                        'Volume':'SPHealth_Volume'})
SP500_Health

##Discretionary
SP500_Discretionary =SP500_Discretionary.drop(columns=['High','Low', 'Adj Close'])
SP500_Discretionary= SP500_Discretionary.rename(columns={'Open': 'SPDisc_Open',
                        'Close':'SPDisc_Close',
                        'Volume':'SPDisc_Volume'})
```

```python
### Create combined SP500 
Combined_SP500= SP500_Material.merge(SP500_Energy, how='left')
Combined_SP500= Combined_SP500.merge(SP500_Finance, how='left')
Combined_SP500= Combined_SP500.merge(SP500_Industrial, how='left')
Combined_SP500= Combined_SP500.merge(SP500_Tech, how='left')
Combined_SP500= Combined_SP500.merge(SP500_Staples, how='left')
Combined_SP500= Combined_SP500.merge(SP500_Utilities, how='left')
Combined_SP500= Combined_SP500.merge(SP500_Health, how='left')
Combined_SP500= Combined_SP500.merge(SP500_Discretionary, how='left')

Combined_SP500
```

```python
### Import the data 

VIX = pd.read_csv("Resources/^VIX.csv")
DOWJONES = pd.read_csv("Resources/DIA.csv")
GLD = pd.read_csv("Resources/GLD.csv")
RUSSELL2K = pd.read_csv("Resources/IWM.csv")
NASDAQ = pd.read_csv("Resources/QQQ.csv")
TENYEAR= pd.read_csv("Resources/^TNX.csv")
SPY=pd.read_csv("Resources/SPY.csv")

####Drop columns and update data
## VIX
VIX= VIX.drop(columns=['High','Low', 'Adj Close'])
VIX= VIX.rename(columns={'Open': 'VIX_Open',
                        'Close':'VIX_Close',
                        'Volume':'VIX_Volume'})
## DOW
DOWJONES= DOWJONES.drop(columns=['High','Low', 'Adj Close'])
DOWJONES= DOWJONES.rename(columns={'Open': 'DOW_Open',
                        'Close':'DOW_Close',
                        'Volume':'DOW_Volume'})
## GLD
GLD= GLD.drop(columns=['High','Low', 'Adj Close'])
GLD= GLD.rename(columns={'Open': 'GLD_Open',
                        'Close':'GLD_Close',
                        'Volume':'GLD_Volume'})
## Russell 2k
RUSSELL2K= RUSSELL2K.drop(columns=['High','Low', 'Adj Close'])
RUSSELL2K= RUSSELL2K.rename(columns={'Open': 'R2K_Open',
                        'Close':'R2K_Close',
                        'Volume':'R2K_Volume'})

## NASDAQ
NASDAQ= NASDAQ.drop(columns=['High','Low', 'Adj Close'])
NASDAQ= NASDAQ.rename(columns={'Open': 'NASDAQ_Open',
                        'Close':'NASDAQ_Close',
                        'Volume':'NASDAQ_Volume'})
## 10YR
TENYEAR= TENYEAR.drop(columns=['High','Low', 'Adj Close'])
TENYEAR= TENYEAR.rename(columns={'Open': '10YR_Open',
                        'Close':'10YR_Close',
                        'Volume':'10YR_Volume'})

## SPY

SPY= SPY.drop(columns=['High','Low', 'Adj Close'])
SPY= SPY.rename(columns={'Open': 'SPY_Open',
                        'Close':'SPY_Close',
                        'Volume':'SPY_Volume'})
```

```python
### Create Combined DF VIX  DOWJONES GLD RUSSELL2K NASQAD  TENYEAR SPY

Combined_Variables=VIX.merge(DOWJONES, how='left')
Combined_Variables= Combined_Variables.merge(GLD, how='left')
Combined_Variables= Combined_Variables.merge(RUSSELL2K, how='left')
Combined_Variables= Combined_Variables.merge(NASDAQ, how='left')
Combined_Variables= Combined_Variables.merge(TENYEAR, how='left')
Combined_Variables= Combined_Variables.merge(SPY, how='left')
Combined_Variables
```
